```python
# Spécifications des âges et des proportions
n = len(df)  # Nombre de lignes dans le DataFrame

# Générer les âges selon les proportions spécifiées
ages_24_40 = np.random.randint(24, 41, size=int(n * 0.60))  # 60% entre [24, 40]
ages_18_24 = np.random.randint(18, 25, size=int(n * 0.10))  # 10% entre [18, 24]
ages_41_67 = np.random.randint(41, 68, size=int(n * 0.26))  # 26% entre [41, 67]
ages_68_70 = np.random.randint(68, 71, size=int(n * 0.03))  # 3% entre [68, 70]

# Vérifier la somme des âges
total_ages = np.concatenate([ages_24_40, ages_18_24, ages_41_67, ages_68_70])

# Ajuster le tableau d'âges si nécessaire
if len(total_ages) != n:
    total_ages = np.resize(total_ages, n)

# Mélanger les âges pour obtenir une répartition aléatoire
np.random.shuffle(total_ages)

# Ajouter la colonne 'age' au DataFrame
df['age'] = total_ages

# Afficher les premières lignes du DataFrame
print(df.head())
```
